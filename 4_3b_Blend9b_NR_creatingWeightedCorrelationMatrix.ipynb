# Used to get the pixel counts

```python
array_dict = {}
water_pixels = {}
for i in range(len(corr_set)):
  # Read in our csv.
  corr_csv = pd.read_csv(f"{src_file}/{subdirectories[i]}/{corr_set[i]}",converters = {'correlation_list':eval})
  # Grab out our correlation matrix.
  corr_mat = np.array(corr_csv['correlation_list'].tolist()[0])
  # Grab out our pixel value.
  pixel_val = corr_csv['pixel_count'][0]

  # Pass both of these values to their respective dictionaries.
  array_dict[f'{corr_set[i]}'] = corr_mat
  water_pixels[f'{corr_set[i]}'] = pixel_val
```

```python
max_pixel = max(water_pixels.values())
```

```python
water_pixels['max_pixel_value'] = max_pixel
```

```python
pd.Series(water_pixels).to_csv('/content/gdrive/My Drive/s2_pixelValues.csv')
```

# Used for the weighted correlation matrix

```python
# This is the total number of water pixels.
total_w = sum(water_pixels.values())

weighted_arrays = {}
# For valid pixels in water_pixels
for water in water_pixels:
  if water == 'max_pixel_value':
    pass
  else:
    # What fraction does each set of pixels account for the total?
    weight = water_pixels[water] / total_w
    # Apply that fraction to each weight's respective correlation matrix.
    weighted_arrays[water] = array_dict[water] * weight
```
